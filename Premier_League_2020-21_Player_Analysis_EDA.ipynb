From the results above and in the histogram and boxplots, we can see that the average age of players in the premier league is 25.5 years old. We can also see from the age distribution of each individual club within the Premier League that Manchester United have the youngest average squad and Crystal Palace have the oldest average squad,

However, from my domain knowledge of the Premier League, I know that this information can be slighlty misleading. For example, although Manchester United have the lowest average_age of appearance makers across the Premier League, my hypothesis is that they do not actually rely on young players as much as some other teams. Some players in the Manchester United squad are youth players who only played a single game towards the end of last season when results didn't matter, hence did not actually have a significant contribution to the team or the final position of the team in the league.

A better analysis to see which clubs relied on young players the most would be to see the average age of players on the pitch per minute per club. Let's analyse this:

```python
# Create a new column in the dataframe
df['Mins_times_age'] = df['Mins'] * df['Age']
df.head()
```

```python
# create an average age reliance score per club
club_list = list(df['Club'].unique())
age_reliance = []
for i in club_list:
        age_minutes = (df[df['Club'] == i]['Mins_times_age'].sum()) / (df[df['Club'] == i]['Mins'].sum())
        x = (i, age_minutes)
        age_reliance.append(x)
```

```python
age_reliance_df = pd.DataFrame(age_reliance, columns=['Club', 'age_reliance'])
```

```python
# show the teams with the lowest average age_reliance
age_reliance_df.sort_values(by='age_reliance', ascending=True).head()
```

```python
# merge the average age and age_reliance together into one dataframe
age_reliance_df2 = pd.merge(age_reliance_df, df.groupby(by='Club').mean()['Age'], on='Club')
age_reliance_df2.columns = ['Club', 'age_reliance', 'average_age']
age_reliance_df2.sort_values(by='age_reliance')
```

```python
# create a bar plot analysing the difference between average age and age_reliance
age_reliance_df2.sort_values(by='age_reliance').plot(x='Club', y=['age_reliance', 'average_age'], kind='bar',
                     figsize=(16,6), title='Average Age and Age reliance for each Club in Premier League', ylabel='Age')
plt.show()
```

This proves my hypothesis as correct, the age_reliance is the average age of a player in each team per minute on the pitch. As we can see, although Manchester United had the lowest average age in terms of players who made appearances during the season, Aston Villa is actually the club which relied most heavily on young players throughout the season. The average age per minute for Aston Villa is 24.5. 

As we can see, for the most part, the age_reliance is higher than the average_age with only 4 clubs breaking the trend.

We can clearly see that Crystal Palace having a very aging squad as they rank highest in both average_age and age_reliance

## Players who control the game

For this analysis, I want to assess which players occupy the ball and are most involved in play when they are on the pitch. The players who have made the most passes p90 can be seen as players who occupy the ball most frequently and control football matches. 

From my domain knowledge, I would expect these players to be players to top teams as they possess most of the possession and most likely to be deep-lying midfielders or defenders as these are the players who pass the ball around in front of the opposition waiting for the opportunity to pass the ball to wingers or the strikers.

```python
# create a new column which assesses the number of passes attempted p90
def passes_p90(passes):
    return (passes['Passes_Attempted']/passes['Mins'])*90

df['Passes_Attempted_p90'] = (df['Passes_Attempted'] / df['Mins']) *90
```

```python
df['Passes_Attempted_p90'] = df['Passes_Attempted_p90'].round(2)
df['Passes_Attempted_p90'].head()
```

```python
#df.sort_values(by='Passes_Attempted_p90', ascending=False) 
top_passes_p90 = df[df['Mins'] > 500].sort_values(by='Passes_Attempted_p90', ascending=False).head(20)
plt.figure(figsize=(12, 6))
plt.xticks(rotation=70, fontsize=8)
sns.barplot(x='Name', y='Passes_Attempted_p90', data=top_passes_p90, palette='viridis')
plt.title('Top 20 Players with largest number Passes_Attempted_P90 (Players with more than 500 minutes)')
plt.tight_layout()
```

This analysis proves my hypothesis correct, the highest number of passes attempted p90 seem to be deep-lying midfielders and defenders for top clubs and these are the players which seem to have most control of the game.

Interestingly, 18 players from the top 20 are from Liverpool, Manchester City and Chelsea, three of the four clubs to finish in the top 4 and all these three clubs are renowned for having large proportions of possession in matches.

## Top goal-scorers and Top-assisters

Although taking penalties are a skill in itself, it is generally regarded that a penalty is a much easier opportunity to score than an open play goal. Therefore, it is interesting and important to analyse Premier League top goal-scorers without taking into account penalties. It is interesting to see the different names that appear on the barplot when disregrding penalties

```python
df['Goals_no_penalties_p90'] = (df['Goals_no_penalties'] / df['Mins']) * 90
plt.figure(figsize=(16,6))
sns.barplot(data=df[df['Goals'] > 2].sort_values(by='Goals_no_penalties_p90', ascending=False).head(20),x='Name',
            y='Goals_no_penalties_p90', palette='viridis')
plt.xticks(rotation=60)
plt.title('Top 20 goalscorers per 90 in the Premier League (no penalties & more than 2 goals)')
plt.show()
```

This data is extremely interesting to see as it gives a goal to game ratio excluding penalties and as we can see, Harry Kane does drop further down in this. We can see that Gareth Bale is the only player in the league to register a goal-to-game ratio better than 1 in this metric. Other players in this list may come as a surprise making it an extremely interesting analysis to see.

```python
plt.figure(figsize=(16,6))
sns.barplot(data=df.sort_values(by='Assists', ascending=False).head(20),x='Name',
            y='Assists', palette='viridis')
plt.xticks(rotation=60)
plt.title('Top 20 assist makers in the Premier League')
plt.show()
```

We see that Harry Kane tops both lists for goals and assists showing what a crucial and important player he is. The likes of Bruno Fernandes, Son Heung-Min, Marcus Rashford and Raheem Sterling also place fairly highly in both lists proving their importance in games

## Expected Goals/ Expected Assists and Unluckiest Players

```python
df['xG_Total'] = (df['xG'] * (df['Mins']/90)).round(2)
df['xA_Total'] = (df['xA'] * (df['Mins']/90)).round(2)
```

```python
plt.figure(figsize=(16,4))
sns.barplot(data=df.sort_values(by='xG_Total', ascending=False).head(10), x='Name', y='xG_Total')
plt.xticks(rotation=45)
plt.title('Players with the highest Expected Goals')
plt.show()
```

```python
plt.figure(figsize=(16,4))
sns.barplot(data=df.sort_values(by='xA_Total', ascending=False).head(10), x='Name', y='xA_Total')
plt.xticks(rotation=45)
plt.title('Players with the highest Expected Assits')
plt.show()
```

In the plot of expected goals and expected assists, we can see that this does in fact roughly match up with the plots for actual top goalscorers and assits makers 

```python
df['Goals_xG_Diff'] = df['Goals'] - df['xG_Total']
highest_GxG_diff = df.sort_values(by='Goals_xG_Diff').head(20)
highest_GxG_diff.plot(x='Name', y=['xG_Total', 'Goals'], kind='bar', figsize=(16,6),
                     title='Players with the Highest Difference between Expected Goals and Goals (Negative)')
```

The above plot shows the players with the highest difference between Expected Goals and Goals. These players effectively underperformed and should have scored quite a number of more goals than they actually did. I have labelled this plot as unlucky players however, there is more at play here than luck for example, they could be missing chances they should have scored due to poor finishing attributes on the player's part and from the top list of players we can see that. The likes of Timo Werner, Neal Maupay, Roberto Firmino and Anthony Martial were poor finishers last season meaning they missed easy opportunities and hence why this led to a large discrepancy between expected goals and actual goals.

```python
highest_GxG_diff = df.sort_values(by='Goals_xG_Diff', ascending=False).head(20)
highest_GxG_diff.plot(x='Name', y=['xG_Total', 'Goals'], kind='bar', figsize=(16,6),
                     title='Players with the Highest Difference between Expected Goals and Goals (Positive)')
```

The above plot shows the top players who outperformed their expected goals. In a similar vein, these players could either be considered lucky for scoring more than they should or can be regarded as clinical finishers for scoring a higher proportion of more difficult chances.

```python
print('The total number of Goals in the Premier League 20/21: {} \nThe total number of Expected Goals in the Premier League 20/21: {}'.format(df['Goals'].sum(), df['xG_Total'].sum().round(2)))
```

In general we can say that expected goals is a good measure of goals as the difference between the total number of goals and total number of expected goals is minimal

## Most aggressive players

```python
# player with the most red cards
df[df['Red_Cards'] == df['Red_Cards'].max()]
```

```python
# player with the most yellow cards
df[df['Yellow_Cards'] == df['Yellow_Cards'].max()][['Name', 'Yellow_Cards']]
```

```python
# top 5 players with most yellow cards
df.sort_values(by='Yellow_Cards', ascending=False)[['Name', 'Club', 'Yellow_Cards']].head()
```

```python
# create new column (aggressiveness score) which gives total cards per player per minute with greater weighting to red cards
df['Aggresiveness_Score'] = (df['Yellow_Cards'] + (df['Red_Cards']*2)) / df['Mins']
```

```python
# Players with the highest aggresiveness score
df.sort_values(by='Aggresiveness_Score', ascending=False)[['Name', 'Club', 'Red_Cards', 'Yellow_Cards', 'Aggresiveness_Score', 'Mins']].head(10)
```

```python
# As with other statistics it is best to filter with players over approx. 400 mins to give better understanding
df[df['Mins'] > 1000].sort_values(by='Aggresiveness_Score', ascending=False)[['Name', 'Club', 'Red_Cards','Yellow_Cards', 'Aggresiveness_Score', 'Mins']].head(10)
```

The table above shows the top 10 players in the Premier League with the highest aggressiveness score for players with over 1000 minutes in the premier league

```python
df[df['Mins'] > 1000].groupby(by='Nationality')['Aggresiveness_Score'].describe().sort_values(by='mean', ascending=False).head(10)
```

The table above shows the top 10 'most aggressive' nationalities based on the aggresiveness score. Note that there are not enough players for each nationality for this to be a true measure of aggressiveness for each nationality. Switzerland tops the list largely thanks to Granit Xhaka, a well known hot-headed, aggressive player.

```python
df[df['Mins'] > 1000].groupby(by='Position')['Aggresiveness_Score'].describe().sort_values(by='mean', ascending=False)
```

The table above shows the most aggressive positions according to the aggressiveness score. We can see a general trend that midfielders and defenders tend to be the most aggressive players whereas, attackers seem to have a relatively lower aggressiveness score

```python
df[df['Mins'] > 1000].groupby(by='Club')['Aggresiveness_Score'].describe().sort_values(by='mean', ascending=False)
```

This above table lists the 'most aggressive' clubs based on the aggressiveness score. We see Sheffield United as the most aggressive team and Liverpool as the least aggressive team.

There appears to be a negative correlation between the possession a team has and their aggressiveness score. This intuitively makes sense as a team with less of the ball will tend to make more fouls as you can only make a foul on an opponent when you are out of possession. Let's test this:

```python
# create merged dataframe to test correlation
df_corr = pd.merge(df[df['Mins'] > 1000].groupby(by='Club')[['Aggresiveness_Score', 'Passes_Attempted']].describe()['Passes_Attempted']['mean'],
         df[df['Mins'] > 1000].groupby(by='Club')[['Aggresiveness_Score', 'Passes_Attempted']].describe()['Aggresiveness_Score']['mean'],
                   on='Club'
)

df_corr.columns = ['Mean_Passes_Attempted', 'Mean_Aggressiveness_Score']
```

```python
# see results for correlation
df_corr.corr()
```

As can be seen there does seem to be a negative correlation between the average passes attempted/possession of a team and their aggrssiveness score. This does intuitively make sense as a team with less possession have a greater probability of making more fouls.
